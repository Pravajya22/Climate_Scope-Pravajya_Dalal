## Exploratory Data Analysis

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# To display all columns
pd.set_option('display.max_columns', None)
```

Load Dataset

```python
# Load the CSV file
df = pd.read_csv("weather_filled.csv")

# Display first 5 rows
df.head()
```

Basic Exploration

```python
# Dataset shape
print("Shape of dataset:", df.shape)

# Info about columns
print("\nDataset Info:")
df.info()

# Summary statistics
print("\nSummary Statistics:")
df.describe()

# Check column names
print("\nColumns:")
print(df.columns.tolist())
```

Check Missing Values and Duplicates

```python
# Missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Duplicate rows
print("\nNumber of duplicate rows:", df.duplicated().sum())
```

Data Types and Key Variables

```python
# Check data types
print("\nData Types:")
print(df.dtypes)
```

Univariate Analysis

```python
# Distribution of numerical columns
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

Bivariate Analysis

```python
# Correlation heatmap
numeric_df = df.select_dtypes(include=[np.number])

# Create a figure with better size and spacing
plt.figure(figsize=(12, 8))
sns.heatmap(
    numeric_df.corr(),
    annot=False,         # Hide correlation numbers (optional)
    cmap='coolwarm',
    linewidths=0.5,      # Add grid lines between cells
    cbar_kws={'shrink': 0.7}
)

plt.title("Correlation Heatmap (Numeric Features)", fontsize=14, pad=12)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()

# Example scatterplots
sns.pairplot(df[numeric_cols[:4]])  # you can limit columns for readability
plt.show()
```

Outlier Detection

```python
# Detect Outliers using Boxplots
numeric_cols = df.select_dtypes(include=[np.number]).columns

for col in numeric_cols:
    plt.figure(figsize=(6, 3))
    sns.boxplot(x=df[col])
    plt.title(f"Outlier Detection for {col}")
    plt.tight_layout()
    plt.show()
```

## Data Preprocessing

Remove Duplicates

```python
df = df.drop_duplicates()
print("Duplicates removed. New shape:", df.shape)
```

Handle missing values

```python
# Fill missing numerical values with median
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

# For categorical columns
cat_cols = df.select_dtypes(exclude=[np.number]).columns
df[cat_cols] = df[cat_cols].fillna(df[cat_cols].mode().iloc[0])

print("Missing values handled.")
```

Convert Data Types

```python
# Example: convert 'Date' column to datetime if exists
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')

print("Data types converted where necessary.")
```

Feature Engineering

```python
if 'Date' in df.columns:
    df['Year'] = df['Date'].dt.year
    df['Month'] = df['Date'].dt.month
    df['Day'] = df['Date'].dt.day

print("New features created.")
```

Save cleaned and preprocessed data

```python
# Save cleaned data
df.to_csv("weather_cleaned.csv", index=False)
print(" Cleaned dataset saved as 'weather_cleaned.csv'")
```
